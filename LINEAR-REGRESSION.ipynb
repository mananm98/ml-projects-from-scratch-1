```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Data

```python
x = np.random.randn(100)
x.shape
```

```python
e = np.random.randn(100)
y = (2*x + 5) + 0.4*e
```

```python
plt.scatter(x,y)
```

```python
def hypothesis(x,theta):
    return theta[0] + theta[1]*x
```

```python
def error(x,y,theta):
    m = x.shape[0]
    e = 0
    for i in range(m):
        hx = hypothesis(x[i],theta)
        e += (hx-y[i])**2
      
    return e
```

```python
def gradient(x,y,theta):
    m = x.shape[0]
    grad = np.zeros((2,))
    for i in range(m):
        hx = hypothesis(x[i],theta)
        grad[0] += 2*(hx-y[i])
        grad[1] += 2*(hx - y[i])*x[i]
       
    return grad
```

```python
def gradient_descent(x,y,learning_rate=0.001,iterations=100):
    theta = np.zeros((2,))
    error_list = []
    m = x.shape[0]
    plt.scatter(x,y)
    for i in range(m):
        g = gradient(x,y,theta)
        theta[0] = theta[0] - learning_rate*g[0]
        theta[1] = theta[1] - learning_rate*g[1]
        error_list.append(error(x,y,theta))
        plt.plot(x,(theta[0] + theta[1]*x))
       
    return theta,error_list
```

```python
plt.figure(figsize = (10,10))
theta,e_list = gradient_descent(x,y)
```
